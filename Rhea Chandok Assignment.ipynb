# 2. Plot 'Argentina-CABA' and 'Argentina-Catamarca' for number of cases values for entire duration to identify the difference.

```python
data
```

```python
caba=data.query("location=='Argentina-CABA'&unit=='cases'")
```

```python
catamarca=data.query("location=='Argentina-Catamarca'&unit=='cases'")
```

```python
caba.value=caba.value.astype(int)
caba.set_index('report_date',inplace=True)
```

```python
catamarca.value=catamarca.value.astype(int)
catamarca.set_index('report_date',inplace=True)
```

```python
l=[caba.value.sum(),catamarca.value.sum()]
```

```python
labels=["Argentina-Caba","Argentine-Catamarca"]
```

```python
plt.bar(labels,l)
plt.ylabel('Value of Cases')
plt.title("Difference between Argentina-CABA and Argentina-Catamarca for value of cases")
```

```python
cab=caba.loc[:,'value']
cat=catamarca.loc[:,'value']

cab.plot(color='r',figsize=(10,5))
cat.plot(color='b',figsize=(10,5))
plt.xlabel('Duration')
plt.ylabel('Value')
plt.title('Argentina-Caba vs Argentina-Catamarca')
```

# 3. Plot how data field values are distributed across the locations 'Argentina-Buenos_Aires', 'Argentina-CABA', 'Argentina-Catamarca'

```python
data
```

```python


aires=data.query("location=='Argentina-Buenos_Aires'")
caba=data.query("location=='Argentina-CABA'")
catamarca=data.query("location=='Argentina-Catamarca'")

aires.value=aires.value.astype(int)
caba.value=caba.value.astype(int)
catamarca.value=catamarca.value.astype(int)

aires.rename(columns={"value":"BuenosAires"},inplace=True)
caba.rename(columns={"value":"CABA"},inplace=True)
catamarca.rename(columns={"value":"Catamarca"},inplace=True)

aires=aires.loc[:,'BuenosAires']
caba=caba.loc[:,'CABA']
catamarca=catamarca.loc[:,'Catamarca']

```

```python
aires.plot(color='r',figsize=(10,5))
caba.plot(color='b',figsize=(10,5))
catamarca.plot(color='g',figsize=(10,5))
plt.title("Data Field Value distribution")
plt.xlabel("duration")
plt.ylabel("data field values")
plt.legend()
```

```python
l=[aires.sum(),catamarca.sum(),caba.sum()]

labels=['Argentina Buenos-Aires','Argentina-Catamarca','Argentina-CABA']
plt.pie(l,labels=labels)
plt.title("Case Values")
```
